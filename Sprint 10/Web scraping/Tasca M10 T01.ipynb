### Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup

URL = "http://quotes.toscrape.com/"
page1 = requests.get(URL)

soup = BeautifulSoup(page1.content, "html.parser")
```

```python
soup
```

```python
quotes = soup.find_all(class_="quote")
```

```python
for q in quotes:
    print(q, end="\n"*2)
```

```python
for q in quotes:
    quote_text = q.find("span", class_="text")
    quote_author = q.find("small", class_="author")
    print(quote_text.text)
    print(quote_author.text)
    print()
```

```python
results2 = driver.find_elements(By.CLASS_NAME, "cursa-content")

for r in results2:
    print('\033[1;3m' + r.find_element(By.CLASS_NAME, "title-link").text + '\033[0m')
    print(r.find_element(By.TAG_NAME, 'p').text)
    links = r.find_elements(By.CLASS_NAME, "main-link.resultat")
    
    for l in links:
        print(l.text + " = " + l.get_attribute("href"))
    print()
```

### Race results details from xipgroc.cat scraped using Scrapy

```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings

class CursasSpider(scrapy.Spider):
    name = "title_link"
    
    def __init__(self, *args, **kwargs):
        self.timeout = int(kwargs.pop('timeout', '60'))
        super(CursasSpider, self).__init__(*args, **kwargs)
        
    def start_requests(self):
        from twisted.internet import reactor
        reactor.callLater(self.timeout, self.stop)    
        
        urls = ['https://xipgroc.cat/cursas']
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        for cursa in response.css(".cursa-content"):
            yield {
                "text": cursa.css(".title-link::text").get()
            }

    def stop(self):
        self.crawler.engine.close_spider(self, 'timeout')
        
```

```python
process = CrawlerProcess(settings={
    "FEEDS": {
        "items.json": {"format": "json"},
    },
})

process.crawl(CursasSpider)
process.start()
```
