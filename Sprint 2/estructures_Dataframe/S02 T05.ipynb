* Create new columns (average flight speed, whether if it arrived late or not...)

```python
df['AvgSpeed'] = df.apply(lambda row: row.Distance / row.ActualElapsedTime, axis=1)
```

```python
import numpy as np

df['Delayed'] = np.where(df['ArrDelay']>0, True, False)
```

* Table of the airlines with the most accumulated delays

```python
print(df['Delayed'].value_counts())
```

```python
df_delays = df.groupby(['UniqueCarrier'])[['Delayed']].sum().sort_values(by='Delayed', ascending=False)
```

```python
df_delays
```

* Which are the longest flights? 

```python
df.sort_values(by='ActualElapsedTime', ascending=False)[["ActualElapsedTime", "UniqueCarrier","FlightNum","Origin","Dest","Distance"]]
```

* And the most delayed ones?

```python
df.sort_values(by='ArrDelay', ascending=False)[["ActualElapsedTime", "UniqueCarrier","FlightNum","Origin","Dest","Distance","ArrDelay","Delayed"]]
```
